```python
#Variables selection for the model
all_wells=well_cut[['EPAssetsId','TVD','Formation','WellTypeStandardised','TotalDepth','Surf_Longitude','Surf_Latitude','KBElevation']]
data_wells=all_wells[all_wells.TVD.notnull()]
data_wells.columns=['EPAssetsId','TVD','Formation','WellTypeStandardised','TotalDepth','Surf_Longitude','Surf_Latitude','KBElevation']
data_wells
```

```python
#Creating dummies for categorical variables WellProfile and WellType
#all_status_dummies = pd.get_dummies(data_wells.CurrentStatus, prefix='Status').iloc[:,1:]
#all_profile_dummies = pd.get_dummies(data_wells.WellProfile, prefix='WellProfile')

all_type_dummies = pd.get_dummies(data_wells.WellTypeStandardised, prefix='WellType')
all_formation_dummies = pd.get_dummies(data_wells.Formation, prefix='Formation')

df_all = pd.concat([data_wells,all_formation_dummies,all_type_dummies], axis=1)
df_all
```

```python
#Extrating feature variables for the model (Normalised varibles)
X1=df_all.iloc[:, 4:]
y1=df_all.iloc[:, 1]
X1
```

```python
#Selecting the random set from the above analysis
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X1_train,X1_test,y1_train,y1_test=train_test_split(X1,y1, random_state=42,test_size=0.3)

scaler = StandardScaler()
X1_train = scaler.fit_transform(X1_train)
X1_test  = scaler.transform(X1_test)
```

```python
print('Training Features Shape:', X1_train.shape)
print('Training Labels Shape:', y1_train.shape)
print('Testing Features Shape:', X1_test.shape)
print('Testing Labels Shape:', y1_test.shape)
```

# Note

Usually it's a good practice to apply following formula in order to find out the total number of hidden layers needed.

Nh = Ns/(α∗ (Ni + No))

where

Ni = number of input neurons.
No = number of output neurons.
Ns = number of samples in training data set.
α = an arbitrary scaling factor usually 2-10.


# Neural Network

```python
#Import the model

import tensorflow
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from keras.layers import Dense
from keras.layers import Dropout
from keras.callbacks import ModelCheckpoint
from keras.models import Sequential
from keras.layers import Dense, Activation, Flatten
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error 
# Initialising the ANN

model_nn = Sequential()

# The Input Layer:
model_nn.add(Dense(10,input_dim = X1.shape[1], activation='relu'))

# Adding the input layer and the first hidden layer

model_nn.add(Dense(100,activation='relu'))
model_nn.add(Dense(100,activation='relu'))
model_nn.add(Dense(100,activation='relu'))


# Adding the output layer

model_nn.add(Dense(1))

#model.add(Dense(1))
# Compiling the ANN
model_nn.compile(loss='mean_squared_error', optimizer='Nadam', metrics=['mean_squared_error'])
model_nn.summary()
```

```python
checkpoint_name = 'Weights-{epoch:03d}--{val_loss:.5f}.hdf5' 
checkpoint = ModelCheckpoint(checkpoint_name, monitor='val_loss', verbose = 1, save_best_only = True, mode ='auto')
callbacks_list = [checkpoint]
```

```python
# Fitting the ANN to the Training set
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(monitor='val_loss', min_delta=1e-3, patience=5, verbose=1, mode='auto', restore_best_weights=True)

hist=model_nn.fit(X1_train,y1_train, epochs=650, batch_size=120, validation_data=(X1_test,y1_test),callbacks=[checkpoint])
```

```python
print(hist.history.keys())
plt.figure(figsize=(10,5))
# "Loss"
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')

plt.show()
```

```python
# Load wights file of the best model :
wights_file = 'Weights-1881--389.94543.hdf5' # choose the best checkpoint 
model_nn.load_weights(wights_file) # load it
model_nn.compile(loss='mean_absolute_error', optimizer='adam', metrics=['mean_absolute_error'])
```
